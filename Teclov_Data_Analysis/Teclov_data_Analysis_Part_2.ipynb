## Data Cleaning & Imputation

```python
# importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# let's import the datasets

rounds = pd.read_csv('round_final.csv', encoding= "ISO-8859-1")
companies = pd.read_csv('companies_final.csv', sep = '\t', encoding = '"ISO-8859-1"')
```

```python
# now first let's verify the permalinks of both datafile

print(len(rounds['company_permalink'].unique()))
print(len(companies.permalink.unique()))

# let'f find if there is any permalink in rounds but not in companies

print(len(rounds.loc[~rounds['company_permalink'].isin(companies.permalink),:]))
```

# Taking Care of Missing value

```python
# check in rounds dataset
rounds.isnull().sum()
```

```python
# check in companies dataset
companies.isnull().sum()
```

```python
# since there is no missing data in permalink so let's merge the two dataset into one

df = pd.merge(companies, rounds,  how = 'inner', left_on='permalink', right_on='company_permalink')
df.head()
```

```python
# sincee the company permalink appears two times let's remove company_permalink from the table
master = df.drop(['company_permalink'], axis = 1)
```

```python
master.head()
```

```python
# now let's check the columns
master.columns
```

```python
# let's check the missing value
master.isnull().sum()
```

```python
# now find the fraction of the missing values

round(100* (master.isnull().sum()/len(master.index)),2)
```

Clearly, the column funding_round_code is useless (with about 73% missing values). Also, for the business objectives given, the columns homepage_url, founded_at, state_code, region and city need not be used.

```python
master = master.drop(['funding_round_code', 'homepage_url', 'founded_at', 'state_code', 'region', 'city'], axis = 1)
```

```python
master.columns
```

```python
# now let's sum up the missing values & display the fraction

round(100 * (master.isnull().sum()/len(master.index)), 2)
```

```python
# note that the raised_amount_usd which we want to analyze and also we need to use country_code
```

```python
master['raised_amount_usd'].describe()
```

Here the mean is around 10M and the median is 1M
Now let's find the raised_amount_usd which has NaN values and remove them form the list

```python
master = master[~np.isnan(master['raised_amount_usd'])]
round(100*(master.isnull().sum()/len(master.index)), 2)
```

```python
# there are some missing values in countrt code column
# let's find out them
```

```python
master['country_code'].value_counts()
```

```python
# now let's view the fraction of the country code

round(100*(master['country_code'].value_counts()/len(master.index)),2)
```

```python
# since there are very few missing value in country code let's drop them
master = master[~pd.isnull(master['country_code'])]
```

```python
master.head()
```

```python
# let's look at the missing values now
round(100*(master.isnull().sum()/len(master.index)), 2)
```

the category_list has a small fraction of missing values let's remove those

```python
master = master[~pd.isnull(master['category_list'])]
```

```python
# let's check for missing values
round(100*(master.isnull().sum()/len(master.index)),2)
```

```python
# The dataset is cleaned now.
# writing the clean dataframe into a new dataframe
master.to_csv('master_final.csv')
```
